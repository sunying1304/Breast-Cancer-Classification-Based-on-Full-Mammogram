```python
from keras.layers import Convolution2D, MaxPooling2D, Activation
from keras.models import Sequential
import numpy as np
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
#ROC
pic = cv2.imread('/Users/sunying/Desktop/db3/jpg3/0+Calc-Training_P_00005_RIGHT_MLO_1.jpg')
pic=np.array(pic)
plt.imshow(pic)
plt.show()
#full mammogram
pic1 = cv2.imread('/Users/sunying/Desktop/db3/fullimage/train/0+Calc-Training_P_00005_RIGHT_MLO.jpg')
pic1=np.array(pic1)
plt.imshow(pic1)
plt.show()
# a,b,c=pic.shape[0],pic.shape[1],pic.shape[2]
# pic=pic.reshape(c,a,b)
```

```python
model = Sequential()

model.add(Convolution2D(30,    # number of filter layers
                        5,
                        5,
                        input_shape=pic.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2
                                 )))
model.add(Convolution2D(30,    # number of filter layers
                        4,
                        4,
                        input_shape=pic.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2
                                 )))
#datas_placeholder = tf.placeholder(tf.float32, [None, 113, 121, 1])

def visualize_pic(model, pic):
    pic_batch = np.expand_dims(pic,axis=0)
    conv0 = model.predict(pic_batch)
    conv0 = np.squeeze(conv0, axis=0)
    #print(conv0.shape)
    #print(conv0)
    for i in range(30):     
        plt.imshow(conv0[:,:,i])
        plt.show()
        
print("for ROI: ")
visualize_pic(model, pic)

print("#############################################################################################")
print("#############################################################################################")
print("#############################################################################################")
print("for full image: ")
model = Sequential()
model.add(Convolution2D(30,    # number of filter layers
                        5,
                        5,
                        input_shape=pic1.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Convolution2D(30,    # number of filter layers
                        4,
                        4,
                        input_shape=pic.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2
                                 )))

visualize_pic(model, pic1)


```
